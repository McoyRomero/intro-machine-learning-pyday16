## Definiendo ML

 * Programas con parámetros ajustables (=modelo) que se ajustan automáticamente (=algoritmo) para obtener como salida datos conocidos
 * Encontrar patrones, generalizar reglas
 * Y = f(x) <- encontrar f para un X e Y conocidos
 * ...


## Ejemplo


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import seaborn; seaborn.set()


def plot_linear_regression():
    a = 0.5
    b = 1.0

    x = 30 * np.random.random(20)

    y = a * x + b + np.random.normal(size=x.shape)

    clf = LinearRegression()
    clf.fit(x[:, None], y)

    x_new = np.linspace(0, 30, 100)
    y_new = clf.predict(x_new[:, None])

    ax = plt.axes()
    ax.scatter(x, y)
    ax.plot(x_new, y_new)

    ax.set_xlabel('x')
    ax.set_ylabel('y')

    ax.axis('tight')
    
plot_linear_regression()
plt.show()
```
